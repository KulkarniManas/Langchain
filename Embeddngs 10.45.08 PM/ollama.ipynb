## Ollama supports embedding models, and building RAG applications

```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
embeddings = (
    OllamaEmbeddings(model = "gemma:2b") # by default it uses llama2
)
```

```python
embeddings
```

```python
r1 = embeddings.embed_documents(
    [
        "Alpha is the first letter of the greek vocabulary",
        "Beta is the second character of the greek vocabulary"
    ]
)
```

```python
r1[1]
```

```python
len(r1[1])
```

```python
embeddings.embed_query("What is the second character of greek vocabulary?")
```

```python
text = "This is the text"
embeddings.embed_query(text)
```

```python
embeddings.embed_query("What is the text?")
```
