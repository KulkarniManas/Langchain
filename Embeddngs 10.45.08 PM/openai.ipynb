# Embedding:

Converting text into vectors

```python
os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")
```

```python
from langchain_openai import OpenAIEmbeddings
#model
embeddings = OpenAIEmbeddings(model = "text-embedding-3-large")
embeddings
```

```python
# testing the openai embeddings
text = "This is the openai tutorial"
query_result = embeddings.embed_query(text)
query_result
```

```python
len(query_result)
```

```python
# we can also change the dimensions
embeddings_1024 = OpenAIEmbeddings(model = "text-embedding-3-large",dimensions=1024)
# further call the embed_query
```

```python
# Load a document
# Split it into document chunks
# Convert it into vector
# Store it into vector database

from langchain_community.document_loaders import TextLoader
loader = TextLoader("./speech.txt")  
docs = loader.load()  # loaded as document
docs

# convert the document into chunks of text
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size = 500, chunk_overlap = 500)
final_documents = text_splitter.split_documents(docs)  # chunks of documents

# Vector embedding and vector storeDB
from langchain_community.vectorstores import Chroma

db = Chroma.from_documents(final_documents,embeddings_1024)
db


```

```python
# retreiving the results
query = "It is a distressing and oppressive duty"
retreived_results = db.similarity_search(query)
print(retreived_results)
```
